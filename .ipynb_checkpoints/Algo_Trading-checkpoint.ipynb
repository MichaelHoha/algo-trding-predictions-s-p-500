```python
def Show_Graph(data, title, color):
    plt.figure(figsize=(20,10))
    plt.plot(data, color)
    plt.title(title)
    plt.grid()
    plt.show()
```

# Use "^GSPC" for the S&P 500 Index


```python
import yfinance as yf
import pandas_ta as ta
sp500 = yf.Ticker("^GSPC")
hist = sp500.history(period="max")
hist
```

# Retrieve historical data




```python
start_date = "2000-01-01"
end_date = "2023-12-31"

hist = sp500.history(start=start_date,end=end_date)
hist
```

```python
vol_sp500 = hist["Volume"][-1000:-1]
Show_Graph(vol_sp500, "sp500 Trading volume last 1000 days", 'red')
```

```python
low_sp500 = hist["Low"][-1000:-1]
Show_Graph(low_sp500, "sp500 Trading low last 1000 days", 'Blue')
```

```python
high_sp500 = hist["High"][-1000:-1]
Show_Graph(high_sp500, "sp500 Trading high last 1000 days", 'Blue')
```
